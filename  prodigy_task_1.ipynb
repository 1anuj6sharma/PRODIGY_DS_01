```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
df = pd.read_csv('data.csv', skiprows=4,  encoding='utf-8')
df.head()
```

```python
fd_final = df.drop(['Country Code' , 'Indicator Name' , 'Indicator Code'] , axis = 1)
```

```python

chosen_country = 'United States'


country_data = fd_final[fd_final['Country Name'] == chosen_country]


country_data = country_data.loc[:, ~country_data.columns.str.contains('^Unnamed')]

years = country_data.columns[2:].astype(int)

population = country_data.iloc[:, 2:].values.flatten()

plt.figure(figsize=(5, 3))
plt.bar(years, population, color='skyblue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of {chosen_country} Over the Years')


plt.show()


```

```python
plt.hist(population,bins=20, color='pink',edgecolor='black')
plt.show()
```
